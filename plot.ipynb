```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_res(ht1, ht2, ht4):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(fill, ht1)
    ax.plot(fill, ht2)
    ax.plot(fill, ht4)
    ax.legend(["HT1", "HT2", "HT4"])
    ax.set_title("Real/Imag")
    ax.set_xlabel("% of cache line elements adjacent")
    ax.set_ylabel("Speedup")
    fig.canvas.draw()
```
